```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
creditcard_data=pd.read_csv('creditcard.csv')
```

```python
class0_data=creditcard_data.loc[creditcard_data['Class']==0]
```

```python
print('mean class 0 -',class0_data.Amount.mean())
print('min class 0-',class0_data.Amount.min())
print('max class 0-',class0_data.Amount.max())
print('variance class 0-',class0_data.Amount.var())
print('std dev class 0-',np.sqrt(class0_data.Amount.var()))
```

```python
class1_data=creditcard_data.loc[creditcard_data['Class']==1]
```

```python
print('mean class 1 -',class1_data.Amount.mean())
print('min class 1 -',class1_data.Amount.min())
print('max class 1 -',class1_data.Amount.max())
print('variance class 1 -',class1_data.Amount.var())
print('std dev class 1 -',np.sqrt(class1_data.Amount.var()))
```

```python
class1_data.Amount.mean()
```

```python
class1_data.Time.head(5)
```

```python
plt.rcParams['figure.figsize']=(15,15)
```

```python
plt.scatter(class1_data.Time, class1_data.Amount, color='b')
plt.show()
```

#### Range below 25000

```python
print('fraudulent cases below 25000 time units since time=0, ',len(class1_data.loc[class1_data['Time']<25000]))
```

```python
print('normal cases below 25000 time units since time=0, ',len(class0_data.loc[class0_data['Time']<25000]))
print('percentage of fraud, ',len(class1_data.loc[class1_data['Time']<25000])/len(class0_data.loc[class0_data['Time']<25000]))
```

```python
print('fraudulent cases greater than 25000 and below 50000 time units since time=0, ',len(class1_data.loc[(class1_data['Time']>=25000) & (class1_data['Time']<50000)]))
```
